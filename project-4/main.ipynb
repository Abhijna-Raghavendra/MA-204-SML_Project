```python
#Imports
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import LabelEncoder
from sklearn.manifold import Isomap
from sklearn.manifold import TSNE
```

```python
#Load the data
data = pd.read_csv('data/wine_data.csv')
data = data.dropna(how='any')
data.dropna(how='any', inplace=True)
X = data.iloc[:,1:12]
y = data.iloc[:,0]
```

```python
#Isomap
X_isomap = Isomap(n_components=2, max_iter=5000,n_neighbors=7).fit_transform(X)
X_isomap.shape
```

```python
#t-SNE
X_tsne = TSNE(n_components=2, verbose=1, perplexity=26, n_iter=1250).fit_transform(X)
X_tsne.shape
```

```python
#Plotting
colors = np.array(['c','m','b'])

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(10, 5))
ax1.scatter(X_isomap[:,0],X_isomap[:,1], c= colors[y-1])
ax1.set_title('Isomap')
ax2.scatter(X_tsne[:,0],X_tsne[:,1], c=colors[y-1])
ax2.set_title('t-SNE')
```
